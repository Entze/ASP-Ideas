```python
from clingo.symbol import Number, String
from clingo.control import Control
```

```python
def run(axioms):
    evASP = Control("0")
    evASP.add("base", [], axioms)

    evASP.ground([("base", [])])
    
    with evASP.solve(yield_=True) as handle:
        #print(handle)
        for m in handle: 
            print("Answer:\n{}\n".format(m))
            #print(m)
```

```python
time = "time(0..6)."

obs = """
obs(balance(k_Forderung,0),0).
obs(balance(v_Forderung,0),0).
obs(balance(k_Verbindlichkeit,0),0).
obs(balance(v_Verbindlichkeit,0),0).
obs(balance(k_Kasse,100),0).
obs(balance(v_Kasse,0),0).
obs(balance(k_ScriptLager,0),0).
obs(balance(v_ScriptLager,1),0).
"""

occ = """
occ(activate,1).
occ(payment,3).
occ(transfer,6).
"""

obs_rules = """
obs(balance(Acc, Before - Amount),T+1) :- 
    occ(credit(Acc, Amount),T),
    obs(balance(Acc, Before),T).

obs(balance(Acc, Before + Amount),T+1) :- 
    occ(debit(Acc, Amount),T),
    obs(balance(Acc, Before),T).
"""

occ_rules_activate = """
occ(debit(k_Forderung,1), T) :-
    occ(activate, T).
occ(credit(k_Verbindlichkeit,100), T) :-
    occ(activate, T).

occ(debit(v_Forderung,100), T) :-
    occ(activate, T).
occ(credit(v_Verbindlichkeit,1), T) :-
    occ(activate, T).
"""

occ_rules_payment = """
occ(debit(k_Verbindlichkeit,100), T) :-
    occ(payment, T).
occ(credit(k_Kasse,100), T) :-
    occ(payment, T).

occ(credit(v_Forderung,100), T) :-
    occ(payment, T).
occ(debit(v_Kasse,100), T) :-
    occ(payment, T).
"""

occ_rules_transfer = """
occ(debit(k_Kager,1), T) :-
    occ(transfer, T).
occ(credit(k_Forderung,1), T) :-
    occ(transfer, T).

occ(debit(v_Verbindlichkeit,1), T) :-
    occ(transfer, T).
occ(credit(v_Lager,1), T) :-
    occ(transfer, T).

"""

inertia = """
obs(balance(Acc, Amount), T+1) :-
    obs(balance(Acc, Amount), T),
    time(T),
    not occ(debit(Acc,_), T),
    not occ(credit(Acc,_), T).

"""
run(inertia)
```

```python
run(time+obs+occ+obs_rules+occ_rules_activate+occ_rules_payment+occ_rules_transfer+inertia+"#show obs/2.")
```
